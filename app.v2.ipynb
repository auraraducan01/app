```python
df = pd.DataFrame(data, columns = ['name', 'gender', 'likes_women', 'likes_men', 'preference'])
df['match_type'] = df['gender'] + '-' + df['preference']
```

```python
def matching_selection():
    match_types = ['wny', 'myn', 'wyn', 'mny', 'wyy', 'myy']
    match_type = random.choice(match_types)
    return match_type
```

```python
def data_selection(match_type):
    match_for_df = []
    match_with_df = [] 
    
    if match_type =='wny': 
        match_for_df = df[(df['gender'] == 'w') & (df['likes_women'] == 'n') & (df['likes_men'] == 'y')]
        match_with_df = df[(df['gender'] == 'm') & (df['likes_women'] == 'y')]
        return match_for_df, match_with_df
    
    elif match_type == 'myn':
        match_for_df = df[(df['gender'] == 'm') & (df['likes_women'] == 'y') & (df['likes_men'] == 'n')]
        match_with_df = df[(df['gender'] == 'w') & (df['likes_men'] == 'y')]
        return match_for_df, match_with_df
    
    elif match_type == 'wyn':
        match_for_df = df[(df['gender'] == 'w') & (df['likes_women'] == 'y') & (df['likes_men'] == 'n')]
        match_with_df = df[(df['gender'] == 'w') & (df['likes_women'] == 'y')]
        return match_for_df, match_with_df
    
    elif match_type == 'mny':
        match_for_df = df[(df['gender'] == 'm') & (df['likes_women'] == 'n') & (df['likes_men'] == 'y')]
        match_with_df = df[(df['gender'] == 'm') & (df['likes_men'] == 'y')]
        return match_for_df, match_with_df
    
    elif match_type == 'wyy':
        match_for_df = df[(df['gender'] == 'w') & (df['likes_women'] == 'y') & (df['likes_men'] == 'y')]
        match_with_df = df[df['likes_women'] == 'y']
        return match_for_df, match_with_df

    elif match_type == 'myy':
        match_for_df = df[(df['gender'] == 'm') & (df['likes_women'] == 'y') & (df['likes_men'] == 'y')]
        match_with_df = df[df['likes_men'] == 'y']
        return match_for_df, match_with_df
    else:
        pass
    
```

```python
def user_matching(match_for_df, match_with_df):
    n = 5
    matched = []
    match_for = match_for_df['name'].tolist()
    match_with = match_with_df['name'].tolist()
    
    while len(match_for) >= n and len(match_with) >= n:
        batch_1 = match_for[:n]
        batch_2 = [user for user in match_with if user not in batch_1]
        batch_2 = random.sample(batch_2, n)
        for i in range(n):
            matched.append(batch_1[i])
            matched.append(batch_2)
        for i in range(n):
            matched.append(batch_2[i])
            matched.append(batch_1)
        remove_batches = batch_1 + batch_2
        match_for = [user for user in match_for if user not in remove_batches]
        match_with = [user for user in match_with if user not in remove_batches]
    return match_for_df, match_with_df, matched, remove_batches
```

```python
def main():
 
    match_type = matching_selection()
    match_for_df, match_with_df = data_selection(match_type)
    match_for_df, match_with_df, matched, remove_batches = user_matching(match_for_df, match_with_df)
    
    return match_for_df, match_with_df, matched
```
